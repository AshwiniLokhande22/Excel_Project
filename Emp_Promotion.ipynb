22.Is there any correlation between the number of trainings, awards won, and promotion rate using heatmap.

```python
plt.figure(figsize=(8, 6))
sns.heatmap( data=df[['No_of_trainings', 'awards_won', 'is_promoted']].corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation between Trainings, Awards Won, and Promotion Rate')
plt.show()
```

This heatmap will show the correlation between the three variables. A positive correlation (close to 1) means that as one variable increases, the other does too, while a negative correlation (close to -1) means an inverse relationship.








23. Which department has the highest number of promoted employees with a specific education level using scatter plot

24. What is the promotion rate by gender in each department? plotly

```python
bar = px.bar(df, x='Department',y='is_promoted',color='Gender', barmode='group',
             labels={'is_promoted': 'Promotion Rate', 'Department': 'Department'},
             title='Promotion Rate by Gender in Each Department')
bar.show()
```

x: Represents departments.
y: Promotion rate (average is_promoted value for each gender in a department).
color: Differentiates between males and females.
barmode: Groups the bars by gender for each department.
This Plotly chart will be interactive, allowing you to hover over each bar to see detailed information and zoom in/out as needed.
